```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.loadtxt("x.csv", delimiter=",", usecols=range(256))
y = np.loadtxt("y.csv", delimiter=",", usecols=range(256))
rho = np.loadtxt("rho.csv", delimiter=",", usecols=range(256))
rho_u = np.loadtxt("rho_u.csv", delimiter=",", usecols=range(256))
rho_v = np.loadtxt("rho_v.csv", delimiter=",", usecols=range(256))
u = np.loadtxt("u.csv", delimiter=",", usecols=range(256))
v = np.loadtxt("v.csv", delimiter=",", usecols=range(256))
```

```python
a = plt.pcolormesh(x[-11:,:10], y[-11:,:10], rho[-11:,:10])
plt.colorbar(a)
```

```python
a = plt.pcolormesh(x[-11:,:10], y[-11:,:10], rho_u[-11:,:10])
plt.colorbar(a)
```

```python
a = plt.pcolormesh(x[-11:,:10], y[-11:,:10], rho_v[-11:,:10])
plt.colorbar(a)
```

```python
a = plt.pcolormesh(x[-11:,:10], y[-11:,:10], rho_u[-11:,:10]/rho[-11:,:10])
plt.colorbar(a)
```

```python
a = plt.pcolormesh(x[-11:,:10], y[-11:,:10], u[-11:,:10])
plt.colorbar(a)
```
