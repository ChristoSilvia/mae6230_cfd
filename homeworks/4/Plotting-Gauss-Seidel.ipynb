```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a65 = np.loadtxt("error_decrease_65")
a33 = np.loadtxt("error_decrease_33")
a17 = np.loadtxt("error_decrease_17")
```

```python
plt.semilogy(a65, label="nx=64")
plt.semilogy(a33, label="nx=33")
plt.semilogy(a17, label="nx=17")
plt.legend()
plt.ylabel("L2 Error")
plt.xlabel("# of Gauss-Seidel Iterations")
plt.title("Smaller Grids Converge Much Faster")
plt.grid(True)
```
